#### IMPORT NECESSARY LIBRARIES.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

#### LOADING IN THE DATA.

```python
df = pd.read_csv('Dataset_ecommerce.csv') #reading in the data.

df.info() #displaying information about the dataset.
```

#### DATA EXPLORATION.

```python
df.head() #display the first five rows of data.
```

```python
df.describe() #descriptive statistic display.
```

```python
df.isna().sum() #checking for the null values in the dataset.
```

```python
df[df['CustomerID'].isna()] #viewing some of those rows in the customerID column that have null values.
```

```python
#dropping those null values

df = df.dropna()
```

```python
df.info() #checking that there are no longer any null values
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate']) #converting the column to datetime format.
```

#### EXPLORATORY DATA ANALYSIS.

###### ANALYSIS BY COUNTRY.

```python
#which country has the most product sales?
quantity_by_country = df.groupby(['Country'])['Quantity'].sum()

quantity_by_country
```

```python
#changing this to a dataframe
quantity_by_country = quantity_by_country.reset_index()
```

```python
#sorting in descending order

quantity_by_country = quantity_by_country.sort_values('Quantity', ascending = False).reset_index()

quantity_by_country #viewing the new dataframe
```

###### COUNTRIES WITH THE MOST PRODUCT SALES.

```python
#visualising the top 10 countries with the highest product sales

#sorting and selecting top 10 countries by quantity
top_10_countries = quantity_by_country.head(10)

#using seaborn to plot
plt.figure(figsize=(20, 8)) #setting the figure size
top_10_plot = sns.barplot(data = top_10_countries, x = 'Country', y = 'Quantity')

#plot details
plt.title('Top 10 Countries by Quantity Sold')
plt.xticks(rotation = 45)
plt.xlabel('Country')
plt.ylabel('Quantity Sold')
plt.show()
```

###### The scale is too large and the difference is barely noticeable because the numbers are so similar, changing the y-axis scale.

```python
#creating a column to store the divided figures.
top_10_plot = quantity_by_country.sort_values('Quantity', ascending=False).head(10).copy()
top_10_plot['Quantity_Millions'] = top_10_plot['Quantity'] / 1e6

#using seaborn to plot
plt.figure(figsize = (20, 8)) #setting the figure size
sns.barplot(data = top_10_plot, x = 'Country', y = 'Quantity_Millions', palette = 'pastel')

#plot details
plt.title('Top 10 Countries by Quantity Sold (in Millions)')
plt.xticks(rotation = 45)
plt.xlabel('Country')
plt.ylabel('Quantity Sold')
plt.show()
```

###### COUNTRIES WITH THE LEAST PRODUCT SALES.

```python
#visualising the bottom 10 countries with the highest product sales

#sorting and selecting top 10 countries by quantity
bottom_10_countries = quantity_by_country.tail(10)

#using seaborn to plot
plt.figure(figsize=(20, 10)) #setting the figure size
bottom_10_plot = sns.barplot(data = bottom_10_countries, x = 'Country', y = 'Quantity')

#plot details
plt.title('Bottom 10 Countries by Quantity Sold')
plt.xticks(rotation = 45)
plt.xlabel('Country')
plt.ylabel('Quantity Sold')
plt.show()
```

##### ANALYSIS OF COUNTRIES AND NUMBER OF CUSTOMERS

```python
country_to_customer = df.groupby(['Country'])['CustomerID'].nunique().reset_index().sort_values('CustomerID', ascending = False).reset_index()

country_to_customer
```

```python
#visualizing this

#sorting and selecting top 10 countries by quantity
top_10_countries_to_customer = country_to_customer.head(10)

#using seaborn to plot
plt.figure(figsize=(20, 8)) #setting the figure size
top_10_c2c = sns.barplot(data = top_10_countries_to_customer, x = 'Country', y = 'CustomerID')

#plot details
plt.title('Top 10 Countries to Customer')
plt.xticks(rotation = 45)
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.show()
```

###### The country with the highest sales (Togo) is not the country with the most customers (Mozambique). Nigeria maintains the same spot in both plots though.

##### SALES TREND.

```python
#how has the sales been fluctuating over time?

monthly_sales = df.groupby(df['InvoiceDate'].dt.to_period('M'))['Quantity'].sum() #grouping quantity of sales by month.

monthly_sales #viewing
```

```python
#visualizing this using a line plot

plt.figure(figsize = (20, 10))
plt.plot(monthly_sales.index.strftime("%y-%m"), monthly_sales.values, marker = 'o', linestyle = '-')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

##### REVENUE/TOTAL PRICE- CHECKING FOR REVENUE PER COUNTRY

```python
#creating a total price/revenue column

df['Revenue'] = df['Quantity'] * df['UnitPrice']

df.head(5)
```

```python
#group country by revenue
revenue_top_10 = df.groupby('Country')['Revenue'].sum().sort_values(ascending=False).head(10).reset_index()

#plot
plt.figure(figsize=(20, 10))
sns.barplot(data = revenue_top_10, x = 'Country', y = 'Revenue', palette = 'pastel')

plt.title('Top 10 Countries by Total Revenue')
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### MOST PROFITABLE PRODUCTS.

```python
#grouping by product description and sorting
top_products = df.groupby('Description')['Revenue'].sum().sort_values(ascending=False).head(10).reset_index()

#plot 
plt.figure(figsize=(20, 8))
sns.barplot(data = top_products, x = 'Revenue', y = 'Description', palette = 'pastel')

plt.title('Top 10 Most Profitable Products')
plt.xlabel('Revenue')
plt.ylabel('Product Description')
plt.show()
```
