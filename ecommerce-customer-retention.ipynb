### COHORT ANALYSIS.

```python
#viewing the data

df.head(3)
```

```python
def get_month(x):
    return dt.datetime(x.year, x.month, 1)
```

```python
df['InvoiceDate'] = df['InvoiceDate'].apply(get_month)
```

```python
df.head()
```

```python
#function to assign the cohort date for each customer
#the cohort date is the month of the customer's first purchase
def get_cohort_date(column):
    #grouping by CustomerID and get the minimum InvoiceDate
    df["cohort date"] = df.groupby("CustomerID")["InvoiceDate"].transform("min")
    return df["cohort date"]

#applying the function to the dataset to create a new column for cohort date
df["cohort date"] = get_cohort_date(df)

#displaying the updated dataframe.
df
```

```python
#renaming the column.
df.rename(columns={"cohort date": "Cohort Date"}, inplace=True)
```

```python
df.info()
```

#### COHORT INDEX

```python
#define a function to extract the month and year from a datetime column
def get_year_and_month(data, col):
    month = data[col].dt.month  #extract month from the specified column
    year = data[col].dt.year    #extract year from the specified column
    return month, year          #return both as separate Series
```

```python
#apply the function to the "cohort date" column and store results in two variables
first_month, first_year = get_year_and_month(df, "Cohort Date")
```

```python
#display the extracted month values
first_month
```

```python
#display the extracted year values
first_year
```

```python
#apply the function to the "cohort date" column and store results in two variables
latest_month, latest_year = get_year_and_month(df, "InvoiceDate")

latest_month
```

```python
latest_year
```

#### COHORT INDEX.

```python
#define a function to calculate the cohort index
def create_cohort_index(first_month, first_year, latest_month, latest_year):
    year_diff = latest_year - first_year      #calculate difference in years
    month_diff = latest_month - first_month   #calculate difference in months
    index = year_diff * 12 + month_diff + 1   #convert difference to months and add 1
    return index                              #return the cohort index
```

```python
#apply the function to calculate the cohort index for each row
df["cohort_index"] = create_cohort_index(first_month, first_year, latest_month, latest_year)

#displaying the updated dataframe
df
```

#### COHORT TABLE.

```python
#grouping by cohort date and cohort index, then counting unique CustomerID
cohort_info = df.groupby(["Cohort Date", "cohort_index"])["CustomerID"].nunique().reset_index()
```

```python
#renaming column to make it clear
cohort_info.rename(columns = {"CustomerID": "Number of customers"}, inplace = True)

cohort_info
```

```python
#pivoting table to reshape it for cohort analysis
cohort_table = cohort_info.pivot(index = "Cohort Date", columns = ["cohort_index"], values = "Number of customers")

cohort_table
```

```python
#plotting original cohort heatmap with raw numbers
plt.figure(figsize = (20, 20))  #setting figure size
sns.heatmap(cohort_table, annot = True, cmap = "Pastel1", fmt = 'g')  #plotting heatmap with annotations and color map
```

```python
#calculating retention rate by dividing each row by its first value
new_cohort_table = cohort_table.divide(cohort_table.iloc[:, 0], axis = 0)
```

```python
#plotting normalized cohort heatmap (retention rates)
plt.figure(figsize = (20, 10))  #setting figure size
sns.heatmap(new_cohort_table, annot = True, cmap = "Pastel1", fmt = ".0%")  #plotting heatmap with percent format
```

```python
#grouping by cohort date and cohort index, summing quantity
quantity_bought = df.groupby(["Cohort Date", "cohort_index"])["Quantity"].sum().reset_index()

#reshaping data to wide format for heatmap
quantity_table = quantity_bought.pivot(index = "Cohort Date", columns = ["cohort_index"], values = "Quantity")

#formatting cohort date for display
quantity_table.index = quantity_table.index.strftime("%B %Y")

quantity_table
```

```python
#plotting heatmap of total quantity
plt.figure(figsize = (20, 10))  #setting figure size
sns.heatmap(quantity_table, annot = True, cmap = "Pastel1", fmt = ".3f")  #plotting heatmap
```

```python
avg_quantity_bought = df.groupby(["Cohort Date", "cohort_index"])["Quantity"].mean().reset_index()

#reshaping data to wide format for heatmap
avg_quantity_table = avg_quantity_bought.pivot(index = "Cohort Date", columns = ["cohort_index"], values = "Quantity")

#formatting cohort date for display
avg_quantity_table.index = pd.to_datetime(avg_quantity_table.index).strftime("%B %Y")

avg_quantity_table

# #plotting heatmap of average quantity using pastel colors
# plt.figure(figsize = (20, 10))  #setting figure size
# sns.heatmap(average_quantity_table, annot = True, cmap = "Pastel1", fmt = ".3f")  #using pastel colormap
```
